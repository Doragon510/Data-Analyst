# [Bài tập] Phân tích trên bộ dữ liệu Online Retail

```python
import pandas as pd
data = pd.read_csv('data/OnlineRetail.csv', encoding = "ISO-8859-1")
data.info()
```

```python
data.shape
```

```python
data
```

```python
data.pivot_table(values = 'Quantity', index = 'InvoiceNo', columns = 'Country', aggfunc = 'mean')
```

```python
data.pivot_table(values = 'Quantity', index = 'CustomerID', aggfunc = ['min',max])
```

```python
data.pivot_table(values = 'Quantity', index = 'StockCode', aggfunc = ['mean','sum'])
```

```python
data.loc[:,'InvoiceDate']
```

```python
temp = data.loc[:,'InvoiceDate'].str.split(' ')
data.loc[:,'Date']= temp.str[0]
data.loc[:,'Time']= temp.str[1]
data.loc[:,['InvoiceDate','Date','Time']]
```

```python
data.pivot_table(values = 'Quantity', index = 'Date', aggfunc = 'sum')
```
