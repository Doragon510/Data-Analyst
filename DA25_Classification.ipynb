```python
#Demo 1
# example of binary classification task
from numpy import where #Lọc dữ liệu
from collections import Counter #Đếm dữ liệu
from sklearn.datasets import make_blobs #tạo dữ liệu phân nhóm ngẫu nhiên
from matplotlib import pyplot
import seaborn as sns

# define dataset
X, Y = make_blobs(n_samples=200, centers=2, random_state=3)
# summarize dataset shape
print(X.shape, Y.shape)
# summarize observations by class label
counter = Counter(Y)
print(counter)
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y);
```

```python
# summarize first few examples
for i in range(10):
    print(X[i], Y[i])
# plot the dataset and color the by class label
for label, _ in counter.items():
    row_ix = where(Y == label)[0]
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

```python
# from numpy import where
# from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Tạo dataset
X, Y = make_blobs(n_samples=600, centers=2, random_state=3, cluster_std=2)

# sns.scatterplot(X[:,0], X[:,1])
# plt.show()

sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y)
plt.show()
```

```python
import sklearn.model_selection as model_selection
#Chia bộ dữ liệu thành 2 phần train và test
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, train_size=0.6)
# X_train
```

```python
# instantiate the model (using the default parameters)
# logreg = LogisticRegression(random_state=16)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train, Y_train)

#Dự đoán với giá trị đầu vào là X_test
Y_pred = logreg.predict(X_test)

sns.scatterplot(x=X_train[:,0], y=X_train[:,1], hue=Y_train)
plt.show()

sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_test)
plt.show()

sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_pred)
plt.show()
```

```python
#Predict giá trị
import numpy as np
x_ = [5, -2]
x_ = np.array(x_).reshape(1, -1)
# x_
y_ = logreg.predict(x_)
print(y_)
# Biểu diễn điểm này trên đồ thị
sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_test)
sns.scatterplot(x=x_[:,0], y=x_[:,1], color='r')
plt.show()
```

```python
#Evaluate model
conf_m = confusion_matrix(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)

print('conf_m:', conf_m, sep='\n', end='\n\n')
sns.heatmap(conf_m)
plt.show()

print('report:', report, sep='\n')
```
