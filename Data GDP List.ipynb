# [Bài tập] Phân tích trên bộ dữ liệu GDP List

```python
import pandas as pd
data = pd.read_csv('data/GDPlist.csv',encoding = "ISO-8859-1")
data.info()
```

```python
data.rename(columns={'Country':'Nuoc','Continent':'Chauluc','GDP (millions of US$)':'GDP (trieu $)'},inplace=True)
data.head()
```

```python
data.insert(1,'Thanhpho',pd.Series(data.loc[:,'Nuoc'], index=data.index))
data.head()
```

```python
data['Nuoc'].unique()
```

```python
data['Thanhpho'].replace('\xa0AVietnam','Hanoi',inplace=True)
data['Thanhpho'].replace('\xa0Afghanistan','Kabun',inplace=True)
data.head()
```

```python
data.drop(data.loc[data.Chauluc == 'Asia'].index,inplace = True)
data.head()
```

```python
data.drop(data.loc[data['GDP (trieu $)'] < 300000].index,inplace = True)
data.head()
```
