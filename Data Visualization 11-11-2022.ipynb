```python
# khai báo thư viện và set style
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid", palette="hsv")
# load bộ dữ liệu tips có sẵn trong seaborn:
tips = sns.load_dataset("tips")
```

```python
display(tips)
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip');
```

```python
air = sns.load_dataset('flights')
air.info()
display(air)
```

```python
air2 = air.groupby('year')['passengers'].mean()
plt.plot(air2.index,air2);
```

```python
sns.lineplot(data = air, x = 'year', y = 'passengers', estimator = 'mean');
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips);
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips);
```

```python
smoker = tips[tips['smoker']=='Yes']
nosmoker = tips[tips['smoker']=='No']
plt.scatter(smoker['total_bill'], smoker['tip'], color = 'r')
axes1 = plt.gca() 
axes1.scatter(nosmoker['total_bill'], nosmoker['tip'], c= 'b')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('tip')
plt.title('Vi du')
plt.legend()
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', markers = ['o','x'], data = tips);
```
