```python
# Truy xuất tên khách hàng, nước, và thành phố mà nước là Germany, và sắp xếp theo thứ tự ABC tên nước, rồi tên thành phố
query = 'SELECT customerName, country, city FROM customers WHERE (country = "Germany" OR country = "USA") \
ORDER BY country, city ASC'
vd8 = pd.read_sql_query(query,connection)
vd8
```
