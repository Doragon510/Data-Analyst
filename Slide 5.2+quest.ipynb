### Biểu đồ ghép cùng loại trên cùng 1 axes

```python
#Tạo dữ liệu nhân tạo
y1 = np.random.normal(200, 30, size = 100)
y2 = np.random.normal(100, 70, size = 100)
data = pd.DataFrame({'y1': y1, 'y2': y2, 'date': pd.date_range('1/1/2021', periods=100)})
data
```

```python
plt.plot(data.index, data.loc[:,'y1'], 'r', label='y1') #Đường y1 màu đỏ
```

```python
#Khai báo thư viện
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact as interact
print("Khai báo thư viện thành công")
```

```python
#Thay thế giá trị NaN bằng 0
dat = dat.fillna(0)
dat
```

```python
list(dat['Country of Birth'])[:-1]
```

```python
comboBox = widgets.Combobox(value="Vietnam", 
                            description="Home Country",
                            continuous_update = False,
                            options = list(dat['Country of Birth'])[:-1])
display(comboBox)
```

```python
IntSlider = widgets.IntSlider(value=2019,
                              description="Năm",
                             min=1949,
                             max=2019,
                              continuous_update = False,
                             step=1)
display(IntSlider)
```

```python
def func1(country, year):
    temp = dat.loc[dat['Country of Birth']==country,'1949':'2019'] #Lọc ra nước cần tìm, đỡ phải gõ lại nhiều
    plt.plot(np.arange(1949,2020), temp.iloc[0,:])
    plt.title(country)
    ax= plt.gca()
    ax.set_xlabel('Năm')
    ax.set_ylabel('Số người nhập cư')
    ax.scatter(year,dat.loc[dat['Country of Birth']==country, str(year)],color = 'r') #Vẽ biểu đồ điểm
    text_ = temp.loc[:,str(year)].item() #Giá trị của điểm 
    space = 0.05*(temp.max(axis=1) - temp.min(axis=1)) #Shift vị trí (y) của text
    plt.annotate(text_,(int(year)-1, temp.loc[:,str(year)]+space), rotation=90) #Hiển thị gía trị của điểm trên đồ thị
    plt.show()
```

```python
interact(func1, country=comboBox, year=IntSlider);
```
