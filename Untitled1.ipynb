```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print("Khai báo thư viện thành công")
```

```python
x = [2, 4, 6, 8, 10]
y = [12.2, 16.3, 20.5, 25.4, 31.2]
```

```python
# plt.plot(data.loc[:,'x'], data.loc[:,'y'])
plt.plot(x, y)
plt.title('Đồ thị Y theo X',
         fontdict = {'fontsize': 18,
                    'fontweight': 50,
                    'color':'blue',
                    'verticalalignment':'baseline',
                    'horizontalalignment':'center'})
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.show()
```

```python
x1 = np.random.normal(size=10000)
plt.hist(x1, bins = 50)
ax1 = plt.gca()
ax1.set_title('Ví dụ Histogram')
ax1.set_xlabel('Giá trị')
ax1.set_ylabel('Tần suất')
plt.show()
```
